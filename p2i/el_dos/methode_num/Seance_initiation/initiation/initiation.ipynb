## 5. Effectuer des calculs
Les fonctions mathématiques usuelles sont définies dans la bibliothèque standard `math`. Il convient donc de l’importer au préalable :
```Python
import math
```
et de préfixer les noms de fonctions : `math.exp`, `math.sin`...

Pour éviter le préfixe, on peut aussi importer explicitement les fonctions nécessaires :
```Python
from math import exp, sin
```

Le calcul matriciel demande l’importation de la bibliothèque `numpy` (et de la sous-bibliothèque `numpy.linalg` pour le calcul de l’inverse d’une matrice). Il est recommandé d’utiliser le type `array` pour définir une matrice :
```Python
from numpy import array
A = array([[1, 2], [3, 4]])
print(A)
```
Sortie :
```
[[1 2]
 [3 4]]
```
Remarque : la bibliothèque `numpy` contient elle aussi des définitions pour les fonctions usuelles. Elles sont différentes de celles définies dans `math` en ceci qu’elles sont capables d’opérer élément par élément sur des objets de type `array`.

## 6. Tracer la représentation graphique d’une fonction
Pour une fonction d’une variable, il est recommandé d’utiliser la bibliothèque `matplotlib` (plus précisément la sous-bibliothèque `pyplot`) comme dans l’exemple ci-dessous représentant la fonctions sinus sur une période.

````python
from numpy import linspace, pi, sin
from matplotlib.pyplot import figure, gca, xlim, show

fig = figure(figsize=(14, 5))  # Crée une figure de dimensions données.

X = linspace(-pi, pi, 100)  # Crée un tableau X de 100 points régulièrement espacés.
Y = sin(X)  # Crée le tableau image de X par la fonction sinus (importée de numpy).

ax = gca()  # Récupération du repère courant.
ax.plot(X, Y, color='red')  # Trace la fonction.
ax.spines['left'].set_position('zero')  # Axe des ordonnées centré.
ax.spines['right'].set_color('none')  # Pas d’axe à gauche.
ax.spines['bottom'].set_position('zero')  # Axe des abscisses centré.
ax.spines['top'].set_color('none')  # Pas d’axe  en haut.

xlim(-3.5, 3.5)  # Bornes de l’axe des abscisses.

show()
````

Pour une fonction de deux variables, on peut utiliser l’objet `Axes3D` (défini dans `mpl_toolkits.mplot3d`) comme dans l’exemple ci-dessous représentant sur $[-2\,;\,2]\times[-2\,;\,2]$ la fonction $f$ définie par $\displaystyle f(x, y) = \textrm{e}^{-x^2-y^2}$.

````python
from numpy import exp, linspace, meshgrid
from matplotlib.pyplot import figure, cm, show
from mpl_toolkits.mplot3d import Axes3D

fig = figure(figsize=(12, 8))
ax = Axes3D(fig)

# Définition de la fonction à 2 variables.
def f(x, y):
    return exp(-x**2 - y**2)

# Création d'une grille.
X = linspace(-2, 2, 50)
Y = linspace(-2, 2, 50)
X, Y = meshgrid(X, Y)

# Création des valeurs en Z.
Z = f(X, Y)

# Création d’une surface coloriée (en fonction de la valeur en z).
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)
fig.colorbar(surf)

show()
````

### Exercice 7
Écrire un code permettant de représenter sur $[0\,;\,5]$ la fonction $f$ de l’exercice 5.

````python
from numpy import linspace, exp, sin
import time
from matplotlib.pyplot import figure, gca, show, xlim, clf, pause, draw, ion

ion()

for loop in range(1, 100):
    
    X = linspace(0, 5, loop)
    Y = exp(-X)*sin(X**2)

    axes = gca()
    axes.plot(X, Y, color='red')
    axes.spines['left'].set_position('zero')
    axes.spines['bottom'].set_position('zero')
    xlim(0, 5)
    draw()
    pause(0.0001)
    clf()
    
````
