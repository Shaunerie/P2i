```python
import numpy as np
from numpy import pi, cos, exp, log, log10, sqrt
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

#Activité numérique 1
f = lambda x: x**3 - 3*x + 1
β1 = 2*cos(10*pi/9)
β2 = 2*cos(4*pi/9)
β3 = 2*cos(2*pi/9)
print(β1, β2, β3)

fig, ax = plt.subplots(figsize=(10,10))
X = np.linspace(-3, 3, 1000)
plt.grid()
ax.plot(X, f(X))
ax.scatter([β1, β2, β3], [0, 0, 0], color=['red', 'blue', 'purple'], zorder=100)
plt.text(β1, 1, "β1")
plt.text(β2, 1, "β2")
plt.text(β3, 1, "β3")

print(fsolve(f,-2))
print(fsolve(f,0))
print(fsolve(f,2))

phi = lambda x: (x**3 +1)/3 #fonction point fixe de f
print(β1, β2, β3)
print(phi(β1), phi(β2), phi(β3))

ax.plot(X, phi(X))


def pointfixe(phi, α, tol, nmax):
    u = np.array([α])
    n = 0; test = tol +1
    while (test>tol) and (n<nmax):
        u = np.append(u,phi(u[n]))
        test = abs((u[n+1]-u[n])/u[n+1])
        n += 1
    ntol = n
    if n == nmax:
        print("Attention, garde-fou atteint!")
    return u, ntol, test

# Représentation des itérés
def graph_pointfixe(u,N):
    for n in range(0,N-1):
        ax.scatter(u[n],0,marker='o',color='red',s = 50)
        ax.plot([u[n],u[n+1]],[u[n+1],u[n+1]],
                color='red',linewidth=2)
        ax.plot([u[n+1],u[n+1]],[u[n+1],u[n+2]],
                color='red',linewidth=2)
        ax.plot([u[n],u[n]],[u[n+1],0.0],
                color='red',linestyle='--')
```
